```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 # ------------------ Reading ------------------

df=pd.read_csv('student_data.csv',encoding='latin1')
df
```

```python
df.info()
```

```python
 # ------------------ Dropping ------------------

df=df.drop(['Unnamed: 0'],axis=1)
```

```python
 # ------------------ Filling ------------------

df['EthnicGroup']=df['EthnicGroup'].fillna('group C')
df['ParentEduc']=df['ParentEduc'].fillna('some college')
df['ParentMaritalStatus']=df['ParentMaritalStatus'].fillna('married')
df['PracticeSport']=df['PracticeSport'].fillna('sometimes')
df['IsFirstChild']=df['IsFirstChild'].fillna('yes')
df['TransportMeans']=df['TransportMeans'].fillna('school_bus')
df['WklyStudyHours']=df['WklyStudyHours'].fillna('5 - 10')
```

```python
 # ------------------ Imputation ------------------


df['TestPrep']=df['TestPrep'].fillna('none')
```

```python
 # ------------------ Deduplication ------------------

df.drop_duplicates(inplace=True)
```

```python
 # ------------------ Reindexing ------------------

df=df.reset_index()
df
```

```python
 # ------------------ Cleaning ------------------

df.drop('index',axis=1,inplace=True)
df
```

```python
 # ------------------ Create TotalScore ------------------


df['TotalScore'] = df['MathScore'] + df['ReadingScore'] + df['WritingScore']
df
```

```python
df.info()
```

```python
# ------------------ Outlier Detection ------------------

q1=df['TotalScore'].quantile(0.25)
q2=df['TotalScore'].quantile(0.50)
q3=df['TotalScore'].quantile(0.75)
iqr=q3-q1

lb=q1-(1.5*iqr)
ub=q3+(1.5*iqr)
print(lb)
print(ub)

out=df['TotalScore'][(df['TotalScore']<lb) | (df['TotalScore']>ub)]
out

df['TotalScore']=df['TotalScore'].clip(lb,ub)
```

```python
sns.set_style('ticks')

# ------------------ Figure ------------------
fig=plt.figure(figsize=(17,12),facecolor='thistle')
plt.gcf().suptitle('Student Performance Analysis',color='indigo',fontsize=20,fontweight='bold')

# ------------------ Plot 1 ------------------
plt.subplot(2,2,1)
bar=sns.barplot(x='Gender',y='TotalScore',data=df,hue='Gender',palette=['#b39ddb', '#9575cd'],errorbar=None,estimator='sum')
for i in bar.containers:
  bar.bar_label(i,fontsize=10,rotation=0,color='white',label_type='center',fontweight='bold')
plt.gca().set_facecolor('lavender')
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.title('Total Score by Gender',color='indigo',fontsize=13,x=0.5,y=1)

# ------------------ Plot 2 ------------------
plt.subplot(2,2,2)
sns.boxplot(x='TotalScore',data=df,color='#9575cd',width=0.4)
plt.gca().set_facecolor('lavender')
plt.title('Gender Disturibution',color='indigo',fontsize=13,x=0.5,y=1)

# ------------------ Plot 3 ------------------
plt.subplot(2,2,3)
bar=sns.barplot(x='ParentEduc',y='TotalScore',data=df,hue='Gender',palette = ['#ce93d8', '#a38cd6'],errorbar=None,estimator='sum')
for i in bar.containers:
  bar.bar_label(i,fontsize=8,rotation=90,color='white',label_type='center',fontweight='bold')
plt.gca().set_facecolor('lavender')
plt.xticks(rotation=90)
plt.title('Total Score by Parent Education',color='indigo',fontsize=13,x=0.5,y=1)

# ------------------ Plot 4 ------------------
plt.subplot(2,2,4)
bar=sns.barplot(x='TestPrep',y='TotalScore',data=df,hue='Gender',palette =  ['#b39ddb','#ba68c8' ],errorbar=None,estimator='sum')
for i in bar.containers:
  bar.bar_label(i,fontsize=9,rotation=0,color='white',fontweight='bold',label_type='center')
plt.gca().set_facecolor('lavender')
plt.xticks(rotation=90)
plt.title('Total Score by Test Preparation',color='indigo',fontsize=13,x=0.5,y=1)

# ------------------ Layout ------------------
plt.tight_layout(rect=[0, 0, 1, 0.95])
fig.savefig('student_data1.jpg',dpi=500)
plt.show()
```

ð **Conclusion**

The analysis highlights clear patterns in student academic performance across gender, parental background, and test preparation. Overall, **female students demonstrate a slightly higher total score than male students**, indicating marginally better aggregate performance. The gender distribution shows a relatively balanced spread of scores, with no extreme disparities.

Parental education plays a significant role in student outcomes. Students whose parents have **some college education or higher qualifications** tend to achieve higher total scores compared to those from lower educational backgrounds. This suggests that parental academic exposure positively influences student performance.

Test preparation also shows a strong impact. Students who **did not complete test preparation courses collectively scored higher**, which may indicate that naturally higher-performing students are less reliant on structured preparation, while those who completed preparation may be using it as academic support.

Overall, the dashboard demonstrates that **student performance is influenced by a combination of demographic and educational background factors**, with parental education and gender contributing noticeably to variations in total scores.

Sports participation also shows a positive relationship with academic performance. Students who **participate in sports occasionally or regularly** achieve higher total scores than those who never participate, suggesting that balanced physical activity may support academic outcomes.

Weekly study hours demonstrate an optimal pattern, where students studying **5â10 hours per week** attain the highest total scores. Both lower (<5 hours) and excessive (>10 hours) study durations are associated with comparatively reduced performance.
